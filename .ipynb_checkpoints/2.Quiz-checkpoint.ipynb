### 결측데이터 치환

```python
#데이터 불러오기
```

```python
df = pd.read_csv('data/bicycle.csv')
df
```

```python
#결측치 확인
```

```python
df.isnull().sum()
```

```python
#모든 결측치를 0으로 치환
```

```python
df.fillna(0)
```

```python
#모든 결측치를 이전 값으로 치환
```

```python
df.fillna(method='pad')
```

```python
#이동거리 결측치 열을 이동거리 평균으로 치환
```

```python
df['이용거리'].fillna(df['이용거리'].mean(), inplace=True)
df.head()
```

```python
#데이터 불러오기
```

```python
df = pd.read_csv('data/bicycle.csv')
df
```

```python
#이용거리를 최빈값으로 치환
```

```python
print("데이터 개수 : ",df['이용거리'].value_counts())
print("=================================")
print("최빈값 : ",df['이용거리'].mode())
df['이용거리'].fillna(df['이용거리'].mode()[0], inplace=True)
```

```python
#결측 개수 확인, 이용거리 결측값이 없어야 함
```

```python
print("데이터 행, 열 : ",df.shape)
print(df.isnull().sum())
```
